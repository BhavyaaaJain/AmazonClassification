```python
import pandas as pd
import numpy as np
import gensim
```

```python
df = pd.read_csv("ecommerce.csv")
```

```python
df
```

```python
df.columns = ['target', 'desc']
```

```python
df.dtypes
```

```python
df=df.dropna()
df
```

```python
# import gensim
```

```python
text = df.desc.apply(gensim.utils.simple_preprocess)
```

```python
text
```

```python
model = gensim.models.Word2Vec(
    window=5,
    min_count=2,
    workers=2
)
```

```python
model.build_vocab(text)
```

```python
model.train(text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save('./amazon.model')
```

```python
pattern=r'[^a-zA-Z0-9\s]'
df['desc'] = df['desc'].str.replace(pattern, " ", regex=True)
```

```python
pattern=r'[\d]'
df['desc'] = df['desc'].str.replace(pattern, '#', regex=True)
```

```python
df
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
df['desc'] = df['desc'].apply(lambda x: word_tokenize(str(x)))
```

```python
df
```

```python
def get_word_embeddings(tokens):
    word_embeddings = []
    for token in tokens:
        if token in model.wv:
            word_embeddings.append(model.wv[token])
    return word_embeddings

```

```python
df['embeddings'] = df['desc'].apply(get_word_embeddings)
```

```python
df
```

```python
import numpy as np
```

```python
df['avg_embeddings'] = df['embeddings'].apply(lambda x: np.mean(x, axis=0) if len(x)>0 else None)
```

```python
df
```

```python
# df.to_csv('newdf.csv', sep=',', index=False, encoding='utf-8')
```

```python
embed = get_word_embeddings(['phone'])
embed = np.array(embed)
embed
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df.avg_embeddings.shape
```

```python
embed = embed.reshape(1, -1)
embed
```

```python
df.avg_embeddings[0].shape
```

```python
embed.shape
```

```python
def calc_sim(row, embed):
    if row is None:
        return None
    row_embed = row.reshape(1, -1)
    sim = cosine_similarity(row_embed, embed)
    return sim[0][0]
```

```python
df['cosine_similarity'] = df['avg_embeddings'].apply(calc_sim, embed=embed)
```

```python
df
```

```python
df_sorted = df.sort_values(by='cosine_similarity', ascending=False)
```

```python
top_N = 5  # You can change this to the number of top rows you want to retrieve
top_similar_rows = df_sorted.head(top_N)

# Print the top similar rows
print("Top Similar Rows:")
print(top_similar_rows)
```
